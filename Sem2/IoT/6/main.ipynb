```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

```

```python

# Load the CSV files
test_df = pd.read_csv('test.csv', delimiter='\t')
train_1_df = pd.read_csv('train_1.csv', delimiter='\t')
train_2_df = pd.read_csv('train_2.csv', delimiter='\t')

# Display the first few rows of each dataframe to verify
print("Test DataFrame:")
print(test_df.head())

print("\nTrain 1 DataFrame:")
print(train_1_df.head())

print("\nTrain 2 DataFrame:")
print(train_2_df.head())
```

```python
train_df = pd.concat([train_1_df, train_2_df], ignore_index=True)
```

```python
train_df['dstport'] = train_df['tcp.dstport'].fillna(train_df['udp.dstport'])
test_df['dstport'] = test_df['tcp.dstport'].fillna(test_df['udp.dstport'])

print(train_df[['tcp.dstport', 'udp.dstport', 'dstport']].head())
print(test_df[['tcp.dstport', 'udp.dstport', 'dstport']].head())

train_df['srcport'] = train_df['tcp.srcport'].fillna(train_df['udp.srcport'])
test_df['srcport'] = test_df['tcp.srcport'].fillna(test_df['udp.srcport'])

print(train_df[['tcp.srcport', 'udp.srcport', 'srcport']].head())
print(test_df[['tcp.srcport', 'udp.srcport', 'srcport']].head())

# train_df['tcp.len'] = train_df['tcp.len'].fillna(-1)
# test_df['tcp.len'] = test_df['tcp.len'].fillna(-1)

# train_df['udp.length'] = train_df['udp.length'].fillna(-1)
# test_df['udp.length'] = test_df['udp.length'].fillna(-1)

train_df['len'] = train_df['tcp.len'].fillna(train_df['udp.length'])
test_df['len'] = test_df['tcp.len'].fillna(test_df['udp.length'])

print(train_df[['tcp.len', 'udp.length', 'len']].head())
print(test_df[['tcp.len', 'udp.length', 'len']].head())

train_df['transfer.protocol'] = np.where(train_df['tcp.srcport'].notna() & train_df['tcp.dstport'].notna(), 0, 1)
test_df['transfer.protocol'] = np.where(test_df['tcp.srcport'].notna() & test_df['tcp.dstport'].notna(), 0, 1)

print(train_df[['tcp.srcport', 'tcp.dstport', 'udp.srcport', 'udp.dstport', 'transfer.protocol']].head())
print(test_df[['tcp.srcport', 'tcp.dstport', 'udp.srcport', 'udp.dstport', 'transfer.protocol']].head())

train_df['tcp.ack'] = train_df['tcp.ack'].fillna(0)
test_df['tcp.ack'] = test_df['tcp.ack'].fillna(0)

print(train_df[['tcp.ack']].head())
print(test_df[['tcp.ack']].head())

```

```python
# Function to print column properties
def print_column_properties(df):
    for column in df.columns:
        print(f"Column: {column}")
        print(f"  Data Type: {df[column].dtype}")
        print(f"  Number of Null Values: {df[column].isnull().sum()}")
        print(f"  Value Counts (Top 5):")
        print(df[column].value_counts().head(5))

        print("-" * 40)

# Print properties for train_df
print("Train DataFrame Column Properties:")
print_column_properties(train_df)
```

```python
# Calculate the number of null values for each column
null_counts = train_df.isnull().sum()

# Sort columns by the number of null values in ascending order
sorted_columns = null_counts.sort_values().index

top_columns = ['dev', 'ip.version', 'ip.ttl', 'ip.flags.df', 'ip.flags.mf', 'dstport', 'srcport', 'transfer.protocol', 'len', 'tcp.ack']
# top_columns = ['dev', 'ip.version', 'ip.ttl', 'ip.flags.df', 'ip.flags.mf', 'dstport', 'srcport', 'transfer.protocol', 'tcp.len', 'udp.length', 'tcp.ack']
# top_columns = sorted_columns[:11]


train_df = train_df[top_columns]
test_df = test_df[top_columns]

print(train_df)
```

```python
train_df = train_df.dropna(how='all', subset=train_df.columns.difference(['dev']))
test_df = test_df.dropna(how='all', subset=test_df.columns.difference(['dev']))
```

```python
train_df = train_df.dropna()
test_df = test_df.dropna()
```

```python
print(train_df)
```

```python
# Separate features and target variable
X_train = train_df.drop('dev', axis=1)
y_train = train_df['dev']

X_test = test_df.drop('dev', axis=1)
y_test = test_df['dev']
```

```python
# Standardize the features

# Drop rows where ip.version is equal to '4,4'
X_train = X_train[X_train["ip.version"] != '4,4']
X_test = X_test[X_test["ip.version"] != '4,4']
y_train = y_train[X_train.index]
y_test = y_test[X_test.index]

# Parse ip.dsfield hex string to numeric values
try:
    X_train["ip.dsfield"] = X_train["ip.dsfield"].apply(lambda x: int(x, 16))
    X_test["ip.dsfield"] = X_test["ip.dsfield"].apply(lambda x: int(x, 16))
except: pass

# Parse ip.flags.df and ip.flags.mf to boolean values
try:
    X_train["ip.flags.df"] = X_train["ip.flags.df"].apply(lambda x: bool(x))
    X_test["ip.flags.df"] = X_test["ip.flags.df"].apply(lambda x: bool(x))
except: pass

try:
    X_train["ip.flags.mf"] = X_train["ip.flags.mf"].apply(lambda x: bool(x))
    X_test["ip.flags.mf"] = X_test["ip.flags.mf"].apply(lambda x: bool(x))
except: pass

try:
    X_train["tcp.flags.fin"] = X_train["tcp.flags.fin"].apply(lambda x: bool(x))
    X_test["tcp.flags.fin"] = X_test["tcp.flags.fin"].apply(lambda x: bool(x))
except: pass

# Parse object to float values
X_train = X_train.apply(pd.to_numeric, errors='coerce')
X_test = X_test.apply(pd.to_numeric, errors='coerce')


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

# Code to find best parameters for RandomForestClassifier

'''
# Initialize the model
# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the model
rf = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print(f"Best Parameters: {best_params}")

# Initialize the model with the best parameters
model = RandomForestClassifier(**best_params, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Predict on the training set
y_pred_train = model.predict(X_train)

# Predict on the test set
y_pred_test = model.predict(X_test)

# Calculate accuracy
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Training Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
'''
```

```python
best_parameters = {
    'max_depth': 20,
    'min_samples_leaf': 1,
    'min_samples_split': 10,
    'n_estimators': 200
}

model = RandomForestClassifier(**best_parameters, random_state=42)
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Training Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```
