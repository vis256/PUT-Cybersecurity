# Exercise 1 - Histogram


A histogram provides an approximate illustration of how numerical data is distributed. The term was originally introduced by Karl Pearson. To construct a histogram, the initial step involves "binning" or grouping the range of values, essentially dividing the entire value range into a series of intervals, and then tallying the number of values that fall within each interval. These intervals, known as bins, are typically defined as successive, non-overlapping ranges of a variable. The bins must be adjacent and are frequently (although not necessarily) of equal size. An example of a histogram is shown in Figure 1. The x-axis shows the ranges of values, while the y-axis shows the number of occurrences of numbers in a given range.

# Exercise 3 - Expected value, mode, median and standard deviation
For data obtained from different measurements, the mean, mode and median can be determined. To calculate that parameters we should use statistics library.

The expected value is the arithmetic mean of a large number of independently selected outcomes of a random variable. 

The median is the middle value of the data that we have ordered from smallest to largest. 

The mode is the value that occurs most frequently in the data set.

Standard deviation - the classic measure of variability, next to the arithmetic mean, the most widely used statistical concept. Intuitively speaking, the standard deviation tells you how widely the values of some quantity (for example, age, inflation, exchange rate) are scattered around its mean.

```python
import statistics
import os
def print_parameters(values, title):
    mean = statistics.mean(values)
    median = statistics.median(values)
    mode = statistics.mode(values)
    stdev = statistics.stdev(values)

    print(f"\n{title}")
    print(f"{'Parameter':<20}{'Value':<20}")
    print(f"{'-'*40}")
    print(f"{'Mean':<20}{mean:<20}")
    print(f"{'Median':<20}{median:<20}")
    print(f"{'Mode':<20}{mode:<20}")
    print(f"{'Standard Deviation':<20}{stdev:<20}")
    draw(values, 20)

def get_all_data_files():
    return [f for f in os.listdir(os.getcwd()) if f.endswith('.txt')]

# Calculate the remaining parameters for the parameter_value data and for the dataset you created
print_parameters(parameter_values, "parameter_values")

# Calculate the parameters for the dataset created by yourself
print_parameters(data, "my_data")

# Calculate the parameters for the dataset read from files indicated by your instructor.
for data_file_txt in get_all_data_files():
    values = get_values_from_file(data_file_txt)
    print_parameters(values, data_file_txt)
```

# Exercise  4 - Generation of data with a given distribution


The numpy library can be used to generate numbers with a given distribution (import numpy as np).
- uniform distribution: x = np.random.uniform(low, high, size) - (low - Lower boundary of the output interval, high - Upper boundary of the output interval, size - number of samples)
- normal distribution: x = np.random.normal(loc, sd, size) - loc - mean (‚Äúcentre‚Äù) of the distribution, sd - Standard deviation (spread or ‚Äúwidth‚Äù) of the distribution, size - number of samples.  
- exponential distribution: x =  np.random.exponential(scale, size) - scale - distribution parameter,  size - number of samples.

1 Generate sample sets of numbers with a given distribution and parameters (uniform, normal and expotential). Prepare histograms for different intervals. 

In the cell below is an example of how to generate data according to an exponential distribution with parameters 0.1 and 1000

```python
import numpy as np
x =  np.random.exponential(0.1, 10000)
plt.hist(x,100)
```

```python
# Generate sample sets of numbers with a given distribution and parameters (uniform, normal and expotential). Prepare histograms for different intervals. 

# Generate uniform distribution data
uniform_data = np.random.uniform(low=0, high=10, size=1000)
print_parameters(uniform_data, "Uniform Distribution")

# Generate normal distribution data
normal_data = np.random.normal(loc=5, scale=2, size=1000)
print_parameters(normal_data, "Normal Distribution")

# Generate exponential distribution data
exponential_data = np.random.exponential(scale=1, size=1000)
print_parameters(exponential_data, "Exponential Distribution")
```
