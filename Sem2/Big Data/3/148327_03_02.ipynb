Correlation means the relation between variables. Correlation analysis is used to "catch" whether there is a relation between two variables (properties, characteristics). What does correlation mean?  Here it is necessary to understand the relation as a kind of similarity in the "behavior of two characteristics". When one feature, characteristic increases does the other also increase? Or does it decrease? Or does it not change at all?

When calculating the correlation relation between two variables, we get a value whose absolute value (in some correlation coefficients we get both negative and positive values) is called the strength of the correlation, the strength of the relation between two variables. Very often there is a doubt how to interpret the size of such a coefficient, does it mean a strong or weak relationship?

Classification according to J.Guilford:

$|\rho|$=0 - no correlation

$0.0<|\rho|≤0.1$ - weak correlation

$0.1<|\rho|≤0.3$ - weak correlation 

$0.3<|\rho|≤0.5$ - average correlation

$0.5<|\rho|≤0.7$ - high correlation

$0.7<|\rho|≤0.9$ - very high correlation

$0.9<|\rho|<1.0$ - almost full correlation

$|\rho|=1$ - full correlation

Peterson's definition is the most commonly used:

According to this definition, the correlation coefficient $\rho(x,y)$ between two verialbles $x$ and $y$ is calculated as follows:

$$\rho(x,y)=\frac{cov(x,y)}{\sigma_x*\sigma_y}$$,

where:
- $cov(x,y)$ is covariance between $x$ and $y$
$$cov(x,y)=E(x∗y)−(E(x)∗E(y))$$
- E(x) expected value of x
- $\sigma_x$ standard deviation for varialble $x$

## How to plot data?:

```python
import matplotlib

import matplotlib.pyplot as plt

#data display

matplotlib.style.use('ggplot') # set up the style of graph

#setting axis labels
 
plt.xlabel('x')

plt.ylabel('y')

#scatter is a function of the Matplotlib library that is used to generate scatter plots in two dimensions.

#the parameters of the function are the values of x, y and colour 

plt.scatter(x, y)

plt.show()
```

```python
import numpy as np

def calculate_correlation(x, y):
    x = np.array(x)
    y = np.array(y)
    covariance = np.mean(x * y) - np.mean(x) * np.mean(y)
    correlation = covariance / (np.std(x) * np.std(y))
    return correlation

correlation = calculate_correlation(x, y)
print("Calculated correlation coefficient:", correlation)
```

### Correlation coefficient can be calculated with the help numpy library

```python
import numpy as np
correlation = np.corrcoef(x, y)[0,1]
print("The correlation coeffient of the x and y inputs are: \n" ,correlation)
```

```python
#code
def generate_positive_correlation():
    X = np.random.rand(10)
    Y = X + np.random.normal(0, 0.1, 10)
    return X, Y

X, Y = generate_positive_correlation()
correlation = np.corrcoef(X, Y)[0,1]
print(correlation)

def generate_negative_correlation():
    X = np.random.rand(10)
    Y = -X + np.random.normal(0, 0.1, 10)
    return X, Y

X, Y = generate_negative_correlation()
correlation = np.corrcoef(X, Y)[0,1]
print(correlation)
```

```python
def change_quarter_to_numeric_value(quarter_str):
    # quarter_str is 20XX-QX
    # convert it to 20XX * 10 + X
    year, quarter = quarter_str.split('-Q')
    return int(year) * 10 + int(quarter)
```

```python
import pandas as pd
import numpy as np

cell_phone_users_df = pd.read_csv('data/cell_phone_users_in_poland.csv', sep=';')
cell_phone_users_X = cell_phone_users_df['Year']
cell_phone_users_Y = cell_phone_users_df['Number of active SIM cards in millions']

mobile_data_usage_df = pd.read_csv('data/mobile_data_usage_singapore.csv', sep=',')
mobile_data_usage_X = mobile_data_usage_df['quarter']
mobile_data_usage_X = mobile_data_usage_X.apply(change_quarter_to_numeric_value)
mobile_data_usage_Y = mobile_data_usage_df['volume_of_mobile_data']

telecommunications_cost_df = pd.read_csv('data/telecommunication_paid_of_households_by_year.csv', sep=',')
telecommunications_cost_X = telecommunications_cost_df['TIME_PERIOD']
telecommunications_cost_Y = telecommunications_cost_df['OBS_VALUE']
# Filter out where unit is CLV10_MEUR
telecommunications_cost_X = telecommunications_cost_X[telecommunications_cost_df['unit'] == 'CLV10_MEUR']
telecommunications_cost_Y = telecommunications_cost_Y[telecommunications_cost_df['unit'] == 'CLV10_MEUR']

correlation = np.corrcoef(cell_phone_users_X, cell_phone_users_Y)[0,1]
print("The correlation coefficient of cell phone users in Poland is:", correlation)

correlation = np.corrcoef(mobile_data_usage_X, mobile_data_usage_Y)[0,1]
print("The correlation coefficient of mobile data usage in Singapore is:", correlation)

correlation = np.corrcoef(telecommunications_cost_X, telecommunications_cost_Y)[0,1]
print("The correlation coefficient of telecommunications cost in households is:", correlation)

import matplotlib.pyplot as plt
plt.scatter(cell_phone_users_X, cell_phone_users_Y)
plt.xlabel('Year')
plt.ylabel('Number of active SIM cards in millions')
plt.title('Cell phone users in Poland')
plt.show()

plt.scatter(mobile_data_usage_X, mobile_data_usage_Y)
plt.xlabel('quarter')
plt.ylabel('volume_of_mobile_data')
plt.title('Mobile data usage in Singapore')
plt.show()

plt.scatter(telecommunications_cost_X, telecommunications_cost_Y)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Telecommunications cost in households')
plt.show()

```
